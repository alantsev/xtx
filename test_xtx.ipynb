```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
def xtx_test(N, DIM, l, err):
    X = np.random.normal(0, 1, [N, DIM])
    A = np.random.dirichlet(np.ones(DIM))

    Y = X @ A + err * np.random.normal(0, 1, N)
    
    XTX = X.T @ X
    XTX_n = XTX + l * np.identity(DIM)
    
    fig, axs = plt.subplots(2)
    
    axs[0].imshow(XTX)
    axs[1].imshow(XTX_n)
    
    A_hat_reg = np.linalg.inv(XTX_n) @ X.T @ Y
    A_hat = np.linalg.inv(XTX) @ X.T @ Y
    
    return A, A_hat, A_hat_reg
    
    
```

```python
A, A_hat, A_hat_reg = xtx_test(N=20, DIM=200, l=0, err=0)
print(np.array([A, A_hat, A_hat_reg]).T)
```

```python
A, A_hat, A_hat_reg = xtx_test(N=20, DIM=200, l=0, err=0.1)
print(np.array([A, A_hat, A_hat_reg]).T)
```
